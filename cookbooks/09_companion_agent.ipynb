# Task Definition: Companion AI - Storyline Generator and User Persona Management

### Overview

The Julep AI Companion Project is designed to enhance user interactions with AI companions by:
- Generating detailed storylines for conversations.
- Updating session contexts based on chat history.
- Refining user personas to reflect changes in preferences and experiences.

### Task Tools:

* Task 1: Generate a Storyline for a Companion-User Conversation

- **get_agent**: Retrieve a Julep agent by ID.
- **list_users**: List Julep users using a metadata filter.
- **list_agent_docs**: List documents associated with an agent.
- **create_agent_doc**: Create a new document for an agent.

* Task 2: Update Session Situation Based on Chat History

- **get_julep_session**: Retrieve a Julep session by ID.
- **update_julep_session**: Update the situation of a session.
- **history_julep_session**: List the history of a session.

* Task 3: Update the User Persona (and About Section) Based on Chat History

- **get_user_from_ppid**: Retrieve a user using the user PPID.
- **list_user_docs**: List documents associated with a user.
- **create_user_doc**: Create a new document for a user.
- **history_julep_session**: List the history of a session.
- **update_user**: Update user information.

### Task Input:

- **Agent Information**: Includes agent ID, name, about, and instructions.
- **User Information**: Comprises user ID, name, about, and metadata.
- **Session Details**: Consists of session ID and chat history.
- **User Input**: User messages that drive the conversation and task execution.

### Task Output:

* Task 1:
- **Storyline**: Detailed narrative for the conversation.

* Task 2:
- **Session Situation**: Updated context for the session.

* Task 3:
- **User Persona**: Refined user profile based on chat history.

### Task Flows

* Task 1: Generate a Storyline

This task involves creating a detailed narrative for a companion-user conversation. The storyline is structured in sections such as introduction, rising action, climax, falling action, and conclusion.

The task flow consists of:
1. Retrieving agent details using `get_agent` tool with the agent ID
2. Getting user information via `list_users` tool filtered by user PPID
3. Evaluating agent and user data to understand their characteristics
4. Using a specialized prompt to generate the storyline that:
    - Reviews companion's name, about section, and instructions
    - Analyzes user's name and background
    - Creates a structured narrative with clear plot progression
    - Ensures alignment with companion's role and objectives
5. Storing the generated storyline using `create_agent_doc` tool

* Task 2: Update Session Situation

This task updates the session context based on the chat history and user input, ensuring that the AI companion can engage effectively with the user.

The task flow consists of:
1. Retrieving current session details using `get_julep_session` tool with session ID
2. Getting session history via `history_julep_session` tool
3. Evaluating session history and user input to:
    - Track conversation context and emotional state
    - Identify key topics and interests discussed
    - Monitor engagement patterns and preferences
4. Using a specialized prompt to generate a system message that:
    - Provides empathetic and understanding responses
    - Maintains positive communication tone
    - Engages with user interests appropriately
    - Adapts to user's mood and conversation style
    - Respects boundaries and privacy
    - Suggests relevant activities and topics
5. Updating the session situation using `update_julep_session` tool

* Task 3: Update User Persona

This task refines the user persona document to capture changes in demographics, psychographics, and content preferences based on recent chat history.

The task flow consists of:
1. Retrieving user information using `get_user_from_ppid` tool filtered by user PPID
2. Getting existing user persona via `list_user_docs` tool
3. Retrieving session history using `history_julep_session` tool
4. Evaluating user data and chat history to:
    - Track demographic information changes
    - Monitor psychographic preferences
    - Identify content interests and engagement patterns
5. Using a specialized prompt to generate an updated persona that:
    - Updates demographics (age, gender, location)
    - Refines psychographic traits and interests
    - Adjusts content preferences and engagement levels
    - Maintains consistency with previous persona
6. Storing the updated persona using `create_user_doc` tool
7. Updating user's about section via `update_user` tool

```python
import uuid
from julep import Client
import os
import yaml

api_key = os.getenv("JULEP_API_KEY")

# NOTE: these UUIDs are used in order not to use the `create_or_update` methods instead of
# the `create` methods for the sake of not creating new resources every time a cell is run.
AGENT_UUID = str(uuid.uuid4())
TASK_UUID_1 = str(uuid.uuid4())
TASK_UUID_2 = str(uuid.uuid4())
TASK_UUID_3 = str(uuid.uuid4())
USER_ID = str(uuid.uuid4())
SESSION_ID = str(uuid.uuid4())

# Create a Julep client
client = Client(api_key=api_key, environment="dev") # <-- or 'production'
```

### Creating an Agent


Agent is the object to which LLM settings, like model, temperature along with tools are scoped to.

To learn more about the agent, please refer to the [documentation](https://github.com/julep-ai/julep/blob/dev/docs/julep-concepts.md#agent).

```python
# Defining the agent
name = "Rosa"
about = "A girl who's interested in chatting with people. Likes going out and partying in night clubs."
instructions = "Keep your responses short and concise."


# Create the agent
agent = client.agents.create_or_update(
    agent_id=AGENT_UUID,
    name=name,
    about=about,
    instructions=instructions,
    model="gpt-4o",
    default_settings={
        "temperature": 0.5,
        "max_tokens": 1000,
        "top_p": 0.9,
        "frequency_penalty": 0.0,
        "presence_penalty": 0.0,
    }
)

agent = client.agents.get(agent_id=AGENT_UUID)
```

## Task 1: Generate a story line for a Companion-User conversation

## Defining the Task

Tasks in Julep are Github-Actions-style workflows that define long-running, multi-step actions.

You can use them to conduct complex actions by defining them step-by-step.

To learn more about tasks, please refer to the `Tasks` section in [Julep Concepts](https://github.com/julep-ai/julep/blob/dev/docs/julep-concepts.md#tasks).

```python
# Defining the task to generate a story line for the agent
task_def_1 = yaml.safe_load('''
name: Generate a Story line for the agent

tools:

- name: get_agent
  description: Get a julep agent by id
  type: system
  system:
    resource: agent
    operation: get

# This tool will be used with a metadata_filter to get the user based on 
# the "ppid" that is stored in the metadata, rather than the id that julep uses.
- name: list_users
  description: List julep users
  type: system
  system:
    resource: user
    operation: list

- name: list_agent_docs
  description: List the agent docs
  type: system
  system:
    resource: agent
    subresource: doc
    operation: list

- name: create_agent_doc
  description: Create an agent doc
  type: system
  system:
    resource: agent
    subresource: doc
    operation: create

main:
                            
- tool: get_agent
  arguments:
    agent_id: inputs[0]['agent_id']
                            
- tool: list_users
  arguments:
    limit: "1"
    sort_by: "'created_at'"
    direction: "'desc'"
    metadata_filter:
      ppid: inputs[0]['user_ppid']

- evaluate:
    agent: outputs[0]
    user: _[0]

- prompt:
  - role: system
    content: |-
      You are an expert at creating detailed storylines for conversations that take place between a user and its long-term companion.
      The user will provide you with the following:
      - Companion's name
      - Companion's about
      - Companion's instructions
      - User's name
      - User's about

      # Steps to Create the companion's storyline

      1. **Understand the Companion's Context**: Review the provided companion's name, about, and instructions to understand the companion's role, characteristics, objectives, and any specific requirements.

      2. **Outline the Plot**: Create a clear outline of the storyline, ensuring it aligns with the companion's characteristics and instructions.

      3. **Develop the Setting**: Describe the environment where the story takes place, incorporating elements from the companion's description.

      4. **Character Development**: Add depth to the companion and other key characters, focusing on their motivations, persona, and interactions.

      5. **Establish a Conflict**: Introduce a challenge or obstacle related to the companion's instructions that drives the plot forward.

      6. **Resolution**: Craft a resolution for the conflict, showing how the companion uses their skills or experiences personal growth to overcome it.

      7. **Conclude the Story**: Wrap up the story with a closing that resolves any loose ends and reflects on how the companion's journey fulfills their mission or purpose.

      # Output Format

      - The storyline should be written in narrative form. 
      - Break down the story into clear sections such as introduction, rising action, climax, falling action, and conclusion.
      - Use engaging and descriptive language to vividly depict scenes and emotions.

  - role: user
    content: |-
      Companion's name: {{outputs[2]['agent']['name']}}
      Companion's about: {{outputs[2]['agent']['about']}}
      Companion's instructions: {{outputs[2]['agent']['instructions']}}
      User's name: {{outputs[2]['user']['name']}}
      User's about: {{outputs[2]['user']['about']}}
  unwrap: true
                          
- tool: create_agent_doc
  arguments:
    agent_id: inputs[0]['agent_id']
    data:
      title: "'Companion Storyline'"
      content: _
''')

# creating the task object
task_agent_storyline = client.tasks.create_or_update(
    task_id=TASK_UUID_1,
    agent_id=AGENT_UUID,
    **task_def_1,
)

# creating the execution
execution_1 = client.executions.create(
    task_id=TASK_UUID_1,
    input={
      "agent_id": AGENT_UUID,
      "user_ppid": USER_PPID,
    }
)
```

### Monitoring the execution

There are multiple ways to get the execution details and the output:

1. **Get Execution Details**: This method retrieves the details of the execution, including the output of the last transition that took place.

2. **List Transitions**: This method lists all the task steps that have been executed up to this point in time, so the output of a successful execution will be the output of the last transition (first in the transition list as it is in reverse chronological order), which should have a type of `finish`.


<span style="color:olive;">Note: You need to wait for a few seconds for the execution to complete before you can get the final output, so feel free to run the following cells multiple times until you get the final output.</span>


```python
# Get execution details
execution = client.executions.get(execution_1.id)
# Print the output
print("Execution Status: ", execution.status)
print("Execution Current Output: ", execution.output)
```

```python
import json

execution_transitions = client.executions.transitions.list(
    execution_id=execution_1.id).items

for index, transition in enumerate(reversed(execution_transitions)):
    print("Step:", index, ", Type:", transition.type)
    print("output (truncated): ", json.dumps(transition.output, indent=2)[:100])
    print("-" * 100)
```

### Print the generated story line 

```python
# The story line is the output of the second (chronologically, i.e. the
# second transition in the transitions list).
story_line = execution_transitions[1].output
print(story_line)
```

```python
# Defining the task to update the session situation based on the history of the session and the user input
task_def_2 = yaml.safe_load('''
name: Update the session situation based on the history of the session and the user input

tools:

- name: get_julep_session
  description: Get a julep session by id
  type: system
  system:
    resource: session
    operation: get

- name: update_julep_session
  description: Update the session situation
  type: system
  system:
    resource: session
    operation: update

- name: history_julep_session
  description: List the session history
  type: system
  system:
    resource: session
    operation: history

main:

# Get the session
- tool: get_julep_session
  arguments:
    session_id: inputs[0]['session_id']
                            
- evaluate:
    old_situation: _['situation']

# Get the history of the session
- tool: history_julep_session
  arguments:
    session_id: inputs[0]['session_id']
              
# evaluate the history of the session
- evaluate:
    history: list(entry.content[0].text for entry in _.entries)

# based on the history of the sessios and user input, generate a prompt
- prompt:
  - role: system
    content: |-
      You are an AI agent tasked with generating system messages for another agent responsible for human interaction.
      Your role is to analyze the session history and user inputs to craft a SYSTEM message that guides the conversational agent in engaging effectively with the user. 
      Ensure the message supports a welcoming presence, meaningful dialogue, and empathetic responses that align with the user's emotional state.

  - role: user
    content: |-
      The session history is:
      {{_.history}}
      {{'The user has just sent the following message:' + inputs[0].get('user_input') if inputs[0].get('user_input') else ''}}


      Generate a SYSTEM message based on the session history and user input, if any. 
      The output should be a well strucutred and detailed SYSTEM message and nothing else.

      Guidelines for the SYSTEM message:
      1. **Empathy and Understanding**: Respond with empathy, acknowledging the user's feelings and offering comforting words where needed.
      2. **Positive Communication**: Use positive language and a friendly tone to create a welcoming atmosphere.
      3. **Engage with Interests**: Show interest in the user's hobbies, stories, and preferences by asking questions and initiating discussions.
      4. **Adaptability**: Adjust tone and response style based on the user's mood and conversation style.
      5. **Boundaries**: Respect the user's privacy and avoid delving into sensitive topics unless invited.
      6. **Suggestions for Activities**: Suggest activities or discussion topics that could positively impact the user's mood or interests.

      Considerations:
      1. Prioritize creating a safe and supportive environment for the user.
      2. Be mindful of cultural differences and practice inclusivity.
  unwrap: true

- tool: update_julep_session
  arguments:
    session_id: inputs[0]['session_id']
    # The 1247 slicing of the old situation is hardcoded to keep the default situation too
    situation: "outputs[1]['old_situation'][:1247] + NEWLINE + NEWLINE + _"
    # render_templates: "True"
''')

# creating the task object
task_update_session_situation = client.tasks.create_or_update(
    task_id=TASK_UUID_2,
    agent_id=AGENT_UUID,
    **task_def_2,
)

execution_2 = client.executions.create(
    task_id=TASK_UUID_2,
    input={
      "session_id": str(SESSION_ID),
      "user_input": "I'm not really sure. What kind of clubs are you into?",
    }
)
```

```python
# The profiler workflow to update the user persona
task_def_3 = yaml.safe_load(f'''
name: generate_update_persona

# Define the input schema for the workflow
input_schema:
  type: object
  properties:
    user_ppid:
      type: string
    session_id:
      type: string
  required:
    - user_ppid
    - session_id

# Define the tools that are going to be used in the workflow
tools:
- name: get_user_from_ppid
  type: system
  description: Get a user from the user ppid
  system:
    resource: user
    operation: list
  
- name: list_user_docs
  type: system
  description: List the user docs
  system:
    resource: user
    subresource: doc
    operation: list

- name: create_user_doc
  description: Create a user doc
  type: system
  system:
    resource: user
    subresource: doc
    operation: create

- name: history_julep_session
  description: List the session history
  type: system
  system:
    resource: session
    operation: history

- name: update_user
  description: Update the user information
  type: system
  system:
    resource: user
    operation: update

main:
                          
#  get the session history
- tool: history_julep_session
  arguments:
    session_id: inputs[0]['session_id']

#  evaluate the session history
- evaluate:
    session_chat_history: list(entry.content[0].text for entry in _.entries)

# Get the user from the ppid using metadata_filter (returns a list)
- tool: get_user_from_ppid
  arguments:
    limit: "1"
    sort_by: "'created_at'"
    direction: "'desc'"
    metadata_filter:
      ppid: inputs[0]['user_ppid']

# Unwrap the list to get the user
- evaluate:
    user: _[0]

# Get the user persona document using metadata_filter (returns a list)
- tool: list_user_docs
  arguments:
    user_id: _['user']['id']
    limit: "1"
    sort_by: "'created_at'"
    direction: "'desc'"

# Get the doc if it exists
- evaluate:
    doc: _[0] if len(_) > 0 else {{}}

# Get the user persona from the doc if the doc exists
- evaluate:
    user_persona: _['doc'].get('content', "")

# Create the user persona using the prompt step
- prompt:
    - role: user
      content: |
        You are an expert at updating detailed user personas for a user.
        Your task is to update the existing user persona based on the provided user data, session chat history, and past user persona.
        Focus on identifying and highlighting the changes in the user's preferences and experiences.

        #Input#
        Following is the user data:
        User details:
        {{{{outputs[3]['user']['metadata']}}}}
                        
        Session chat history:
        {{{{outputs[1]['session_chat_history']}}}}
        
        User past persona:
        {{{{_.user_persona}}}}

        #Task#
        Carefully review the user data and update the existing user persona by focusing on the following sections:

        1. Demographics
        2. Psychographics
        3. Content Preferences

        For each section, follow these processes:

        1. Demographics:
          - Update the user's age, gender, and location if there are any changes.
          - Make reasonable assumptions for any missing information based on other details provided.

        2. Psychographics:
          - Update the description of the user's personality traits if there are any changes.
          - Modify the interest graph to reflect any new primary and secondary interests based on the session chat history and other relevant areas.

        3. Content Preferences:
          - Update the list of the user's thematic interests if there are any changes.
          - Specify any new content categories that would appeal to this user.
          - Identify any new entity preferences (e.g., specific topics, authors, or genres).
          - Update the user's engagement level with various content types if there are any changes.
          - Highlight any new correlations between the user's interests in various topics and entities.
          - Update the user's content needs and expectations if there are any changes.

        #Instructions#

        When updating the user persona:
        - Be specific and concise.
        - Make logical inferences based on the given information.
        - Ensure consistency throughout the persona.
        - Use bullet points or short paragraphs for easy readability.

        Format your response as follows:
        - Use headers for each main section (Demographics, Psychographics, Content Preferences).
        - Use subheaders for subsections within Content Preferences.
        - Include a brief introduction and conclusion.

        Present the updated user persona within <user_persona> tags, highlighting the changes. Begin with a concise introduction, then proceed with the sections as outlined above.
        Your work will be evaluated for comprehensive and detailed updates to the persona.
  unwrap: true

# Create a new persona document
- tool: create_user_doc
  arguments:
    user_id: outputs[3]['user']['id']
    data:
      title: "'User Persona'"
      content: _

# Update the about section of the user with the new persona
- tool: update_user
  arguments:
    user_id: outputs[3]['user']['id']
    name: outputs[3]['user']['name']
    about: outputs[7]
''')

# creating the task object
task_user_persona = client.tasks.create_or_update(
    task_id=TASK_UUID_3,
    agent_id=AGENT_UUID,
    **task_def_3,
)

execution_3 = client.executions.create(
    task_id=TASK_UUID_3,
    input={
      "user_ppid": USER_PPID,
      "session_id": SESSION_ID,
    }
)
```

```python
execution_transitions = client.executions.transitions.list(
    execution_id=execution_3.id).items

for index, transition in enumerate(reversed(execution_transitions)):
    print("Step:", index, ", Type:", transition.type)
    print("output (truncated): ", json.dumps(transition.output, indent=2)[:100])
    print("-" * 100)
```
