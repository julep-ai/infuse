```python
from vllm import AsyncLLMEngine, AsyncEngineArgs

engine_args = AsyncEngineArgs(
    model="julep-ai/samantha-1-turbo-awq",
    dtype="float16",
    enforce_eager=True,
    max_model_len=8192,
    max_num_seqs=1,
)
```

```python
engine = AsyncLLMEngine.from_engine_args(engine_args)
```

```python
from uuid import uuid4
from vllm.sampling_params import SamplingParams

prompt = "<|im_start|>user\nWhat is the meaning of life?<|im_end|><|im_start|>"

sampling_params = SamplingParams(temperature=0.1, max_tokens=200)
output = engine.generate(prompt, sampling_params, uuid4())
```

```python
async for res in output:
    print(res)
```
