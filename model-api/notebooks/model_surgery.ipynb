## Start engine

```python
from vllm import AsyncLLMEngine, AsyncEngineArgs

args = dict(
    model="julep-ai/samantha-1-turbo",
    tensor_parallel_size=2,
    enforce_eager=True,
    dtype="bfloat16",
)

engine_args = AsyncEngineArgs(**args)
engine_args
```

```python
engine = AsyncLLMEngine.from_engine_args(engine_args)
```

## Prepare generator

```python
from uuid import uuid4
from vllm.sampling_params import SamplingParams

def prep_generator(
    prompt,
    temperature=0,
    max_tokens=200,
    logits_processors=[],
    **sampling_kwargs,
):
    sampling_params = SamplingParams(
        temperature=temperature,
        max_tokens=max_tokens,
        logits_processors=logits_processors,
        **sampling_kwargs,
    )
    
    res_generator = engine.generate(
        prompt,
        sampling_params,
        uuid4(),
    )

    return res_generator

async def generate(
    prompt,
    **sampling_kwargs,
):
    res_generator = prep_generator(prompt, **sampling_kwargs)
    final_res = None

    async for res in res_generator:
        final_res = res
    
    return final_res
```

## Prep logits processor

```python
import torch

tokenizer = engine.engine.tokenizer.tokenizer

identity = lambda x: x
last_requests: tuple[str, list[int], torch.Tensor] = []

def get_lp(prompt, process=identity):
    def processor(
        previously_generated_tokens,
        next_token_logits,
    ):
        last_requests.append(
            (prompt, previously_generated_tokens, next_token_logits.cpu())
        )

        return process(next_token_logits)

    return processor

def reset_requests():
    global last_requests
    last_requests = []
```

## Run all examples

```python
# Add a baseline to examples for comparison
baseline = '<|im_start|>situation\nYou are Samantha. You are talking to Diwank. He is a fun guy.<|im_end|><|im_start|>person (Diwank)\nHi Samantha!<|im_end|>\n<|im_start|>'
example_prompts.insert(0, dict(positive=baseline, negative=baseline))
```

```python
reset_requests()
for example in example_prompts:
    for key in ["positive", "negative"]:
        prompt = example[key]
        logits_processors = [
            get_lp(prompt)
        ]
        
        await generate(prompt, logits_processors=logits_processors, max_tokens=1)
```

```python
# Sanity check
assert len(last_requests) == (len(example_prompts) * 2)
```

## Analyze tags

```python
tags = ["situation", "me", "person", "functions", "function_call", "thought", "information"]

tag_token_ids = [
    tokenizer(tag, add_special_tokens=False)["input_ids"]
    for tag in tags
]

tag_id_map = {
    tag: tag_ids[0]
    for tag, tag_ids in zip(tags, tag_token_ids)
}

id_tag_map = {
    id: tag
    for tag, id in tag_id_map.items()
}
```

```python
from torch.nn import functional as F

get_tag_logits = lambda idx: {
    tag: last_requests[idx][2][id]
    for tag, id in tag_id_map.items()
}

get_tag_probs = lambda idx: {
    tag: F.softmax(last_requests[idx][2], dim = -1)[id]
    for tag, id in tag_id_map.items()
}
```

```python
tag_first_token_ids = list(id_tag_map.keys())
```

```python
def get_dist(idx, upper=50, lower=-2, output_probs=False):
    if output_probs:
        values = F.softmax(last_requests[idx][2], dim=-1)
    else:
        values = last_requests[idx][2]
        
    values = values.tolist()
    
    return [min(upper, max(lower, v)) for v in values]
```

## Visualize

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting
def plot(idx, output_probs=False):
    dist = get_dist(idx, output_probs=output_probs)

    plt.figure(figsize=(20, 6))
    plt.plot(dist, marker='o', linestyle='-', color='blue')
    plt.title(f'Plot of result {idx}')
    plt.xlabel('Index')
    plt.ylabel('Logit')
    
    # Highlighting tags
    # b : blue · g : green · r : red · c : cyan · m : magenta · y : yellow · k : black
    colors = "b,g,r,c,m,y,k".split(',')
    
    for (tag, id), color in zip(tag_id_map.items(), colors):
        plt.axvline(x=id, color=color, linestyle='--', label=tag)  # Indices are 0-based
    
    plt.legend()
    plt.show()

```

```python
show = lambda idx, output_probs=False: (last_requests[idx][0], plot(idx, output_probs))
```
