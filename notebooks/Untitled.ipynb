```python
from memory.cozo import client
from memory.embed import embed
```

```python
text = "What is Diwank like as a person?"
[query_embedding] = embed([text], "query")

n = 100
confidence = 0.5
radius = 1 - confidence

# Embedding search

# query = f"""
# ?[belief, valence, dist, character_id, aspects] := ~beliefs:embedding_space{{ aspects, belief, valence, character_id |
#     query: vec({query_embedding}),
#     k: {n},
#     ef: 20,
#     radius: {radius},
#     bind_distance: dist,
# }}

# :order -valence
# :order dist
# """

keywords = "Diwank person like"
query = f"""
?[belief_id, belief, valence, score, character_id] := ~beliefs:summary {{ belief_id, belief, valence, character_id |
    query: "{keywords}",
    k: {n},
    score_kind: 'tf_idf',
    bind_score: score
}}

:order -valence
:order -score
:limit {n}
"""


client.run(query)
```
